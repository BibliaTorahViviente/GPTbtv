```python
df = pd.read_json(os.path.join("model_judgment", "gpt-4_single.jsonl"), lines=True)
```

```python
pattern = re.compile(r"\[\[(.*?)\]\]")

def get_score(text):
    matches = pattern.findall(text)

    if len(matches) == 1:
        return float(matches[0])
```

```python
get_score(df.iloc[0]["judgment"])
```

```python
scores = []

for i, row in df.iterrows():
    score = get_score(row["judgment"])

    if score:
        scores.append(score)
```

```python
fig = px.histogram(scores)
fig.update_layout(
    xaxis_title="scores given by GPT-4",
    yaxis_title="count",
    height=400,
    width=1000
)
fig.show()
```
