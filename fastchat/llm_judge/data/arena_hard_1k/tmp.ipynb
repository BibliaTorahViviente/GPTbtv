```python
import json

cnt = 0
qid_gpt4_hard = []
for line in open("../../../data/toxicity/data/eval_hard_1k/model_answer/gpt-4.jsonl"):
    obj = json.loads(line)
    if "[[9]]" in obj["choices"][0]["turns"][0]:
        qid_gpt4_hard.append(obj["question_id"])
        cnt += 1
print("gpt-4 hard", cnt)

with open("question.jsonl", "w") as f:
    for line in open("question_shuf.jsonl"):
        obj = json.loads(line)
        
        if obj["question_id"] not in qid_gpt4_hard:
            continue
        
        obj["category"] = "arena_hard_1k"
        obj['turns'] = [obj['turns'][0]]
        f.write(json.dumps(obj) + "\n")
```
